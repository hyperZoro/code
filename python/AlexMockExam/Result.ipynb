```python
import pandas as pd
```

```python
def import_excel(file_path):
    headers = ['id','gender',r'English %',r'Maths %',r'English',r'Maths', r'Total',	r'English_adj',	r'Maths_adj', 	r'Total_adj']
    try:
        df = pd.read_excel(file_path, skiprows =2, header = None, names=headers, dtype={'id': str})
        for col in df.columns:
            if (col != 'id') and (col !='gender'):
                df[col] = pd.to_numeric(df[col], errors='coerce')
        return df
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None
```

```python
file_path = 'data/2025_Test_A_hall-based_results_Excel.xlsx'
df = import_excel(file_path)
```

```python
# Refactored ranking code using loops
subjects = ['Maths', 'English', 'Total']
subjects_adj = ['Maths_adj', 'English_adj', 'Total_adj']

# Regular ranking
for subject in subjects + subjects_adj:
    df[f'{subject}_Rank'] = df[subject].rank(ascending=False)

# Percentage ranking
for subject in subjects + subjects_adj:
    df[f'{subject}_Percentage_Rank'] = df[subject].rank(pct=True, ascending=False) * 100

# Gender-based ranking
for subject in subjects + subjects_adj:
    df[f'{subject}_Rank_Gender'] = df.groupby('gender')[subject].rank(ascending=False)

# Gender-based percentage ranking
for subject in subjects + subjects_adj:
    df[f'{subject}_Gender_Percentage_Rank'] = df.groupby('gender')[subject].rank(pct=True, ascending=False) * 100
```
